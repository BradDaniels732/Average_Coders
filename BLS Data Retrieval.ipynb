```python
#create list to only pull 03(rate) and 06(labor pool)
measure_codes=["03","06"]

#create list to pull chosen area codes
area_codes=["CA3640800000000","CA4828800000000","CA4837200000000","MT4813140000000"]

#initialize dataframe to hold 20 years of state unemployment data
df_unemployment_data = pd.DataFrame(columns=['footnotes', 'latest', 'period', 'periodName', 'value', 'year','area_code', 'measurecode'])

```

```python
# cycle through measure codes

for ms_code in measure_codes:

# cycle through area codes (includes PR & DC)

    for i in range(len(area_codes)):  
        url = f"https://api.bls.gov/publicAPI/v2/timeseries/data/LAU{area_codes[i]}{ms_code}?startyear=2000&endyear=2019&registrationkey={api_key}"
        ud=requests.get(url).json()
        df_unempl_data=pd.DataFrame(ud['Results']['series'][0]['data'])
        df_unempl_data["area_code"] = area_codes[i]
        df_unempl_data["measurecode"] = ms_code
        
        df_unemployment_data=df_unemployment_data.append(df_unempl_data,ignore_index=True, sort=False)
```
